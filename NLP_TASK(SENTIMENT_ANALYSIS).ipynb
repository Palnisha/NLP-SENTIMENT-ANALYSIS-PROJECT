```python
 # Remove non-alphabetic characters
import re
def clean_text(text):
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text
```

```python
## NOW TOKENIZE THE TEXT  FROM NLP BY USING NLTK LIBRARY

from nltk.tokenize import word_tokenize
def tokenize_text(text):
    tokens = word_tokenize(text)
    return tokens
```

```python
## Convert all text to lowercase
def lowercase_text(text):
    return text.lower()
```

```python
## REMOVING STOPWORD
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

def remove_stopwords(tokens):
    filtered_tokens = [word for word in tokens if word not in stop_words]
    return filtered_tokens
```

```python
## STEMMIMNG THE TEXT
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

def stem_words(tokens):
    return [stemmer.stem(word) for word in tokens]
```

```python
#lemmetize the text
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def lemmatize_words(tokens):
    return [lemmatizer.lemmatize(word) for word in tokens]
```

```python
df['cleaned_text'] = df['text'].apply(clean_text)
df['cleaned_text']
```

```python
x = df['cleaned_text']
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
## converting our text data into numerical
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
x_train = vectorizer.fit_transform(x_train)
x_test = vectorizer.transform(x_test)
```

```python
#### train the ML model
## logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression(max_iter = 1000)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test,y_pred)
print('accuracy score on the testing data :' ,accuracy)
```

```python
#### train the ML model
## logistic regression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression(max_iter = 1000)
model.fit(x_train,y_train)
x_train_pred = model.predict(x_train)
accuracy = accuracy_score(y_train,x_train_pred)
print('accuracy score on the training data :' ,accuracy)
```
